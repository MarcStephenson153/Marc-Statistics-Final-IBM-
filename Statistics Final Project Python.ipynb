```python
import pandas as pd
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as pyplot
import scipy.stats
import statsmodels.api as sm
from statsmodels.formula.api import ols 
```

```python
boston_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ST0151EN-SkillsNetwork/labs/boston_housing.csv'
boston_df=pd.read_csv(boston_url)
```

```python
ax = sns.boxplot(y='MEDV', data=boston_df)
ax.set_title("Median Value of Owner Occupied Homes")
```

This is a boxplot for the median value of homes in the $1000's. Where the median is around 20.

```python
ax = sns.barplot(x="CHAS", y="MEDV", data=boston_df)
ax.set_title('Median Value vs Charles River Variable')
```

This is the Median value vs the Charles River variable in which the value is significantly higher when the Charles River variable is present.

```python
ax = sns.boxplot(y='MEDV', x='AGE', data=boston_df)
ax.set_title('Median Value vs Age')
```

```python
ax = sns.scatterplot(x='INDUS', y='NOX', data=boston_df)
ax.set_title("Nitric Oxide Conc vs Non Retail Buisness Acre per Town")
```

There is a postitive correlation between the NOX and INDUS variables.

```python
sns.catplot(x='PTRATIO', kind='count', data=boston_df)
ax.set_title("Student to Teacher Ratio")
```

This is the student to teacher ratio as a histogram showing that there is a town with significant outliers and a higher student to teacher ratio.

Hypothesis: There is a significant difference in median value of houses bounded by the Charles River.



Null Hypothesis: There is no significatn difference in the median value of houses bounded by the Charles River

```python
x=boston_df['MEDV']
y=boston_df['CHAS']
x=sm.add_constant(x)
model=sm.OLS(y,x).fit()
predictions=model.predict(x)
model.summary()
```

The p value score is greater than alpha of 0.05 so we accept the null hypothesis.

Hypothesis: There is a signficance in MEDV prior to AGE 1940
Null Hypothesis: There is no significance in MEDV prior to AGE 1940

```python
x=boston_df['MEDV']
y=boston_df['AGE']
x=sm.add_constant(x)
model=sm.OLS(y,x).fit()
predictions=model.predict(x)
model.summary()
```

The P value is less than alpha of 0.05 therefore we reject the null hypothesis. There is a significance in MEDV of homes built prior to 1940.

Hypothesis: There is a correlation between Nitric Oxcide Concentrations and preportion of non-retail business acres per town.
Null Hypothesis: There is no correlation between Nitric Oxcide Concentrations and preportion of non-retail business acres per town.


```python
x=boston_df['NOX']
y=boston_df['INDUS']
x=sm.add_constant(x)
model=sm.OLS(y,x).fit()
predictions=model.predict(x)
model.summary()
```

Since P is less than 0.05 we can accept the hypothesis that there is a correlation between NOX and INDUS.
